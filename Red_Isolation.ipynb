```python
#Mengimport Library 
import imageio
import matplotlib.pyplot as plt
import numpy as np
a = imageio.imread('bird.jpeg').astype(np.float64)/255

#Mengkestrak warna merah dan menyimpannya pada variable redImg
redImg = a[:,:,0]
#Mengkestrak warna hijau dan menyimpannya pada variable greenImg
greenImg = a[:,:,1]+0.1
#Membagi redImg dengan greenImg
divImg = redImg/greenImg
#Melakukan threshold terhadap divImg dan menyimpannya pada variable rx
rx = (divImg>1)*divImg
#Melakukan normalize pada rx 
rx = rx/rx.max()

#Mengkonversi RGB ke grayscale
#Hasil konversi disimpan di variable grsc
grsc = np.dot(a[...,:3],[0.2126,0.7152,0.0722])
#melakukan Hi Bits dengan n=0xc0, 
bx = (grsc*255.0).astype(np.uint8) & 0xc0
#Ubah Kembali bx jadi Float
bx=(bx.astype(np.float64))/255.0

#Membuat mask dari bx yang memiliki value==0
mx = bx==0

#Menginvers mask
invMx = 1-mx
#Mengalikan invMx dengan bx
bg = invMx*bx
#Kalikan mx dengan rx
#Untuk menyeleksi bagian yang akan berwarna merah
colored = mx*rx	
#Hasil Akhir
finish = np.dstack((colored+bg, 0+bg, 0+bg)) 

#Cetak Hasil Akhir
plt.imshow(finish)



```
