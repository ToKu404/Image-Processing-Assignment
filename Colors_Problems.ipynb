### 5.Tulis skrip yang akan memuat citra uji, ubah menjadi HSV, kalikan nilai H dengan 2/3 dan buat citra RGB baru menggunakan nilai H yang telah dimodifikasi. Tuliskan notasi operator untuk protokol ini.

a. Notasi Operator (*Equation*)
$$\mathrm{\mathbf{a}}[\vec{x}] = Y\left(\mathrm{'bird.jpeg}\right)$$

$$\mathrm{\mathbf{b}}[\vec{x}] = \mathcal{L}_{HSV}\mathrm{\mathbf{a}}[\vec{x}] $$

$$\mathrm{\mathbf{h}}[\vec{x}] =

### 6.Jika konversi warna tanpa kehilangan, maka konversi dari satu ruang dan kemudian kembali ke aslinya akan menghasilkan citra yang sama persis dengan aslinya. Tunjukkan bahwa proses konversi warna tidak kehilangan  beberapa informasi dengan menggunakan operasi berikut pada citra uji $a[\vec{x}]$

$$\mathrm{\mathbf{b}}[\vec{x}] = 
\left|
(\mathcal{L}_{RGB}\mathcal{L}_{YIQ}\mathrm{\mathbf{a}}[\vec{x}]) - \mathrm{\mathbf{a}}[\vec{x}]
\right|
$$

Solusi :


```python
#Load Citra
adata6 = imageio.imread("bird.jpeg")/255

#Konvert warna ke YIQ
adata6_y, adata6_i, adata6_q = rgb2yiq(adata6[...,0],adata6[...,1],adata6[...,2])

#Konvert warna ke RGB
adata6_r, adata6_g, adata6_b = yiq2rgb(adata6_y,adata6_i,adata6_q)
#Buat Citra RGB
adata6_convert = np.dstack((adata6_r, adata6_g, adata6_b))

# Melihat adanyakehilangan informasi dengan mengurangkan data awal dengan data yg telah dikonvert
adata6_out = np.abs(adata6_convert - adata6)

#Tampilkan
plt.imshow(adata6_out)
#Melihat apakah nilai maxnya bukan 0
adata6_out.max()
```

dapat dilihat bahwa citra hasil operasi tampak gelap akan tetapiternyata nilai maximum dari citra bukan 0 yang berarti terdapat kehilangan sejumlah data karena citra sebelum dikonversi ukurannya tidak sama dengan setelah dikonversi.

### 7.Lakukan operasi berikut menggunakan citra uji sebagai  $a[\vec{x}]$ dan simpan jawabannya sebagai image.
$$\mathrm{\mathbf{b}}[\vec{x}] = \mathcal{L}_{HSV}\mathrm{\mathbf{a}}[\vec{x}] $$

$$\mathrm{\mathbf{c}}[\vec{x}] = 
\begin{Bmatrix}
\varnothing \\
1 \\
1
\end{Bmatrix}
\mathrm{\mathbf{b}}[\vec{x}] +
(
\begin{Bmatrix}
1 \\
\varnothing \\
\varnothing
\end{Bmatrix}
\mathrm{\mathbf{b}}[\vec{x}]
)^{1/2}$$

$$\mathrm{\mathbf{d}}[\vec{x}] = \mathcal{L}_{RGB}\mathrm{\mathbf{c}}[\vec{x}] $$

Solusi :

```python
#Citra Uji
adata7 = imageio.imread("bird.jpeg")/255

# Convert adata7 ke HSV
adata7_h, adata7_s, adata7_v = rgb2hsv(adata7[...,0], adata7[...,1], adata7[...,2])
# Pangkat duakan citra dengan channel H
adata7_h_modif = adata7_h**(1/2)
#Convert kembali jadi rgb dan gunakan channel h yang telah dimodif
adata7_r, adata7_g, adata7_b = hsv2rgb(adata7_h_modif, adata7_s, adata7_v)
adata7_output = np.dstack((adata7_r, adata7_g, adata7_b))

#Simpan Citrahasil dalam Image Baru
imageio.imwrite('adata7.jpg', (adata7_output*255).astype(np.uint8))
#Tampilkan output
plt.imshow(adata7_output)
```

### 9.Tuliskan notasi operator untuk metode yang dipilih pada soal sebelumnya.

Solusi :

a. Notasi Operator

Operator untuk memuat citra a$[\vec{x}]$
$$\mathrm{\mathbf{a}}[\vec{x}] = Y\left(\mathrm{"vanandkids.jpg"}\right) $$

Karena Color Model yang dipilih adalah YIQ maka akan dirubah citra a$[\vec{x}]$ menjadi Model YIQ
$$\mathrm{\mathbf{b}}[\vec{x}] = \mathcal{L}_{YIQ}\mathrm{\mathbf{a}}[\vec{x}] $$

Lakukan operator window untuk menyeleksi manual ROI(Fokus Wilayah) yang akan diseleksi (crop)
$$\mathrm{\mathbf{c}}[\vec{x}] = \square_{\vec{v_{1}}, \vec{v_{2}}} $$

Citra yang hanya mengekstrak channel Y
$$\mathrm{\mathbf{d}}[\vec{x}] = 
\begin{Bmatrix}
1 \\
\varnothing \\
\varnothing \\
\end{Bmatrix}\mathrm{\mathbf{c}}[\vec{x}]$$

Citra yang hanya mengekstrak channel I
$$\mathrm{\mathbf{e}}[\vec{x}] = 
\begin{Bmatrix}
\varnothing \\
1 \\
\varnothing \\
\end{Bmatrix}\mathrm{\mathbf{c}}[\vec{x}]$$

Data-data statistic pada ROI

Nilai Minimum
$$
\begin{gather}
m_1 = \bigvee  \mathrm{\mathbf{d}}[\vec{x}] &  m_2 = \bigvee  \mathrm{\mathbf{e}}[\vec{x}]
\end{gather}
$$

Nilai Maximum
$$
\begin{gather}
n_1 = \bigwedge  \mathrm{\mathbf{d}}[\vec{x}] &  n_2 = \bigwedge  \mathrm{\mathbf{e}}[\vec{x}] 
\end{gather}
$$

Mean
$$
\begin{gather}
\mu_1 = \mathcal{M} \mathrm{\mathbf{d}}[\vec{x}] & \mu_2 = \mathcal{M} \mathrm{\mathbf{e}}[\vec{x}]
\end{gather}
$$

Standar Deviasi
$$
\begin{gather*}
\sigma_1 = \mathcal{T} \mathrm{\mathbf{d}}[\vec{x}] & \sigma_2 = \mathcal{T} \mathrm{\mathbf{e}}[\vec{x}]
\end{gather*}
$$

Pengaplikasian  Thresshold
$$\mathrm{\mathbf{f}}[\vec{x}] = 
\begin{pmatrix} 
(\mu_1 - \sigma_1) < \mathrm{\mathbf{d}}[\vec{x}] < (\mu_1 + \sigma_1)
\end{pmatrix}
\times
\begin{pmatrix} 
(\mu_2 - \sigma_2) < \mathrm{\mathbf{e}}[\vec{x}] < (\mu_2 + \sigma_2)
\end{pmatrix}
$$

Melakukan invers karena area van yang akan diisolasi
$$
\mathrm{\mathbf{g}}[\vec{x}] = 1 - \mathrm{\mathbf{f}}[\vec{x}]
$$


b. Script Program Python


```python
# Load Citra
adata9 = imageio.imread("vanandkids.jpeg")/255
plt.imshow(adata9, cmap="gray")
```

```python
# Lakukan Crop Bagian Mobil
adata9_crop = adata9[0:400, 0:950, :]
plt.imshow(adata9_crop)
```

```python
#Persiapan Statistic ROI untuk Y
adata9_y_max = adata9_y.max()
adata9_y_min = adata9_y.min()
adata9_y_mean = adata9_y.mean()
adata9_y_std = adata9_y.std()

print('Min   Y : ', adata9_y_min)
print('Max   Y : ', adata9_y_max)
print('Mean  Y : ', adata9_y_mean)
print('Std   Y : ', adata9_y_std)
```

```python
#Persiapan Statistic ROI untuk I
adata9_i_max = adata9_i.max()
adata9_i_min = adata9_i.min()
adata9_i_mean = adata9_i.mean()
adata9_i_std = adata9_i.std()

print('Min   I : ', adata9_i_min)
print('Max   I : ', adata9_i_max)
print('Mean  I : ', adata9_i_mean)
print('Std   I : ', adata9_i_std)
```

```python
#Lakukan thresshold terhadap Citra Y untuk menyeleksi VAN
adata9_mask_y = np.where(
    (adata9_y_mean - adata9_y_std< adata9_y) & 
    (adata9_y_mean + adata9_y_std> adata9_y), True, False)
plt.imshow(adata9_mask_y, cmap="gray")
```

```python
#Lakukan thresshold terhadap Citra I untuk menyeleksi kids dari van
adata9_mask_i = np.where(
    (adata9_i_mean - adata9_i_std< adata9_i) & 
    (adata9_i_mean + adata9_i_std> adata9_i), True, False)
plt.imshow(adata9_mask_i, cmap="gray")
```

```python
#Kalikan mask y dengan mask q
adata9_mask = adata9_mask_y*adata9_mask_i
# Inverskan
adata9_out = 1-adata9_mask
plt.imshow(adata9_out, cmap="gray")
```
